```python
import pandas as pd

df = pd.read_csv('british_birds_song/birdsong_metadata.csv')

df.head()
```

```python
# Create numeric bird_id from english_cname
df['bird_id'] = pd.Categorical(df['english_cname']).codes
```

```python
df.head(10)
```

```python
import librosa
import pandas as pd
from pathlib import Path

def add_audio_durations(df, audio_directory):
    """Add duration column to dataframe"""
    
    durations = []
    
    for file_id in df['file_id']:
        # Construct filename: xc132608.flac
        audio_file = Path(audio_directory) / f"xc{file_id}.flac"
        
        try:
            # Get duration without loading full audio (faster)
            duration = librosa.get_duration(path=str(audio_file))
            durations.append(round(duration, 2))
        except:
            print(f"Failed to get duration for: xc{file_id}.flac")
            durations.append(None)  # or 0.0
    
    df['duration_seconds'] = durations
    return df

# Usage
df = add_audio_durations(df, "british_birds_song/songs")
```

```python
df.head(10)
```

```python
df_filtered = df.loc[df.groupby('bird_id')['duration_seconds'].idxmax()]
df_filtered.to_csv('british_birds_song/birdsong_metadata_filtered.csv')
```

```python
import librosa
import soundfile as sf
from pathlib import Path

def extract_single_10sec_clip(df, audio_directory, output_directory):
    """Extract single 10-second clip from middle of each recording"""
    
    # Create output directory
    Path(output_directory).mkdir(exist_ok=True)
    
    extracted_files = []
    
    for _, row in df.iterrows():
        file_id = row['file_id']
        bird_id = row['bird_id']
        species_name = row['english_cname']
        
        input_file = Path(audio_directory) / f"xc{file_id}.flac"
        
        try:
            # Load audio
            y, sr = librosa.load(str(input_file), sr=22050)
            
            # Calculate middle portion for 10 seconds
            total_samples = len(y)
            clip_samples = 10 * sr  # 10 seconds worth of samples
            
            # Start position for middle 10 seconds
            start_sample = (total_samples - clip_samples) // 2
            
            # Extract 10-second clip
            clip_start = start_sample
            clip_end = clip_start + clip_samples
            y_clip = y[clip_start:clip_end]
            
            # Save clip
            output_file = Path(output_directory) / f"bird_{bird_id:02d}_xc{file_id}.wav"
            
            sf.write(str(output_file), y_clip, sr)
            
            # Store info
            extracted_files.append({
                'bird_id': bird_id,
                'original_file_id': file_id,
                'clip_path': str(output_file),
                'species_name': species_name,
                'clip_duration': 10.0
            })
            
            print(f"✅ Extracted: bird_{bird_id:02d} ({species_name})")
            
        except Exception as e:
            print(f"❌ Failed: {file_id} - {e}")
    
    return extracted_files

# Usage
clips_info = extract_single_10sec_clip(df_filtered, "british_birds_song/songs", "clips_10sec")

# Save clip info to CSV for reference
import pandas as pd
clips_df = pd.DataFrame(clips_info)
clips_df.to_csv('extracted_clips_info.csv', index=False)
```
